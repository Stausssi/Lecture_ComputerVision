# Segmentation with K-Means
Source taken from https://machinelearningknowledge.ai/image-segmentation-in-python-opencv/

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
img = cv.cvtColor(cv.imread('./img/itsohk_beach.jpg'), cv.COLOR_BGR2RGB)
plt.imshow(img);
plt.title('Original');
plt.show()
```

```python
img_2d = img.reshape((-1, 3))
img_2d = np.float32(img_2d)

# reshape image to be list of 3 channel colors
print(img_2d.shape)
```

```python
# define criteria for k-Means algorithm
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 10, 1.0)
attempts = 10
```

```python
# apply k-means for K=2
K = 2
ret, label, center = cv.kmeans(img_2d, K, None, criteria, attempts, cv.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img.shape))

plt.imshow(result_image);
plt.title('Result with K=2');
plt.show()
```

```python
# apply k-means for K=2
K = 3
ret, label, center = cv.kmeans(img_2d, K, None, criteria, attempts, cv.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img.shape))

plt.imshow(result_image);
plt.title('Result with K=3');
plt.show()
```

```python
K = 4
ret, label, center = cv.kmeans(img_2d, K, None, criteria, attempts, cv.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img.shape))

plt.imshow(result_image);
plt.title('Result with K=4');
plt.show()
```
