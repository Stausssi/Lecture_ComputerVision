# Teste das vorhandene Setup

## Importiere notwendige Bibliotheken

```python
import numpy as np
import pandas as pd
import cv2
from matplotlib import pyplot as plt
import scipy
```

## Lade Bild und zeige es in eigenem Fenster an

```python
img = cv2.imread('./img/lena_std.tif', cv2.IMREAD_COLOR)

showWindow = False
if showWindow:
    cv2.imshow('Testbild Lena, www.lenna.org', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## Lade Bild und zeige es "Inline" an

```python
# Anweisung fr innerhalb
%matplotlib inline
plt.figure()
plt.imshow(img)

# konvertiere Bildfarben
plt.figure()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

## Teste Numpy und SciPy Installation

```python
# noinspection PyUnresolvedReferences
if np.pi == scipy.pi:
    print('Konnte PI aus Numpy und SciPy-Libs auslesen und vergleichen.')
else:
    print('Es gab einen Fehler.')
```

## Bildkoordinatensystem

```python
# Zeige die ersten 25 Spalten des Bildes an
plt.figure()
plt.imshow(img_rgb[:, 0:24])
plt.show()
```

```python
# Zeige die ersten 50 Zeilen des Bildes an
plt.imshow(img_rgb[0:49])
plt.show()
```

## Farbkanle

```python
# Zeige die einzelnen Farbkanle Kanal des Bildes als 3x Grauwert-Bilder an
# Anmerkung: plt.imshow(img[...], cmap='gray', vmin=0, vmax=255) ermglicht die Darstellung als Grauwert-Bild
# und verhindert durch vmin und vmax die automatische Skalierung der Pixelwerte

# Vorher mit BGR2RGB konvertiert -> Reihenfolge RGB
r = img_rgb[:, :, 0]
g = img_rgb[:, :, 1]
b = img_rgb[:, :, 2]
```

### Rot

```python
plt.imshow(r, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Grn

```python
plt.imshow(g, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Blau

```python
plt.imshow(b, cmap='gray', vmin=0, vmax=255)
plt.show()
```
